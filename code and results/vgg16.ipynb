```python
normalize = transforms.Normalize(
   mean=[0.485, 0.456, 0.406],
   std=[0.229, 0.224, 0.225]
)
preprocess = transforms.Compose([
   transforms.Resize(256),
   transforms.CenterCrop(224),
   transforms.ToTensor(),
   normalize
])

# carico modello di classificazione (di tipo vgg)
with open('model.pickle', 'rb') as handle:
    vgg = pickle.load(handle)

# labels
LABELS_URL = 'https://s3.amazonaws.com/outcome-blog/imagenet/labels.json'
response = requests.get(LABELS_URL)
labels = {int(key): value for key, value in response.json().items()}

# leggo immagini nella cartella e scrivo su file i risultati
lista_paesi = os.listdir('images/')
for p in lista_paesi:
    k=0
    lista_immagini = os.listdir('images/'+p+'/')
    f = open('results/'+p+'_results.txt', 'w')
    for i in lista_immagini:
        try:
            img = Variable(preprocess(Image.open('images/'+p+'/'+i)).unsqueeze(0))
            output = vgg(img)
            sm = torch.nn.Softmax(dim=1)
            probs = sm(output)
            prob, j = torch.topk(probs,1)
            prob = (float(prob[0][0]))
            j = int(j[0][0])

            f.write(i+ '\t'+labels[j]+'\t'+str(round(prob,3))+'\n')
        except:
            print(i)
            print('errore')
        k+=1
        if k%10==0:
            print(str(k),end=' ')
    f.close()    
   
```
